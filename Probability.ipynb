```python
#Q1. SIMULATION
#(a) Tossing a coin

import random

trials = 10000
heads = 0
tails = 0

for _ in range(trials):
    toss = random.choice(['H', 'T'])
    if toss == 'H':
        heads += 1
    else:
        tails += 1

print("Experimental Probability of Heads:", heads/trials)
print("Experimental Probability of Tails:", tails/trials)

#We simulated tossing a fair coin 10,000 times using Python’s random module. 
#We counted how many times heads and tails occurred and divided by total trials to estimate experimental probabilities.
```

```python
#(b)Rolling two dice- probability of sum = 7

trials = 10000
sum7_count = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    
    if die1 + die2 == 7:
        sum7_count += 1

print("Experimental Probability of sum = 7:", sum7_count/trials)

#Explanation

#Two dice were rolled repeatedly. We counted how often the sum equals 7 and divided by total trials to estimate the probability.

```

```python
# Q2. At least one “6” in 10 rolls

import random

def prob_at_least_one_six(simulations=10000):
    success = 0
    
    for _ in range(simulations):
        rolls = [random.randint(1,6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    
    return success/simulations

print("Estimated Probability:", prob_at_least_one_six())

#Explanation

# We simulated rolling a die 10 times and checked whether at least one 6 appeared. The proportion of successful trials gives the estimated probability.
```

```python

# Q3. Conditional Probability & Bayes

import random

# Bag composition
balls = ['R']*5 + ['G']*7 + ['B']*8

trials = 1000
prev_blue = 0
red_after_blue = 0

previous = None

for _ in range(trials):
    current = random.choice(balls)
    
    if previous == 'B':
        prev_blue += 1
        if current == 'R':
            red_after_blue += 1
    
    previous = current

conditional_prob = red_after_blue / prev_blue if prev_blue > 0 else 0

print("Estimated P(Red | Previous Blue):", conditional_prob)


#Bayes Verification (Theoretical)


P_red = 5/20
P_blue = 8/20

print("Theoretical P(Red):", P_red)
print("Theoretical P(Blue):", P_blue)

#Explaination

#We simulated drawing balls with replacement and computed the conditional probability from data. 
#Then we compared with theoretical probability to verify Bayes’ theorem.

```

```python
#Q4. Discrete Random Variable


import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.40]

sample = np.random.choice(values, size=1000, p=probabilities)

mean = np.mean(sample)
variance = np.var(sample)
std_dev = np.std(sample)

print("Empirical Mean:", mean)
print("Empirical Variance:", variance)
print("Empirical Standard Deviation:", std_dev)

# Explanation

#We generated 1000 samples using numpy.random.choice according to the given probability distribution. 
#Then we calculated mean, variance, and standard deviation.
```

```python
# Q5. Exponential Distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

mean = 5
scale = mean

data = np.random.exponential(scale=scale, size=2000)

# Histogram
plt.hist(data, bins=30, density=True, alpha=0.6, label="Histogram")

# PDF overlay
x = np.linspace(0, max(data), 100)
plt.plot(x, expon.pdf(x, scale=scale), label="PDF")

plt.title("Exponential Distribution")
plt.legend()
plt.show()

#Explanation

#We generated exponential random samples with mean 5. 
#A histogram was plotted and the theoretical PDF was overlaid for comparison.
```

```python
#Q6. Central Limit Theorem

import numpy as np
import matplotlib.pyplot as plt

# Step a: generate uniform data
uniform_data = np.random.uniform(0, 1, 10000)

# Step b & c: sample means
sample_means = []

for _ in range(1000):
    sample = np.random.choice(uniform_data, size=30)
    sample_means.append(np.mean(sample))

# Plot
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(uniform_data, bins=30)
plt.title("Uniform Distribution")

plt.subplot(1,2,2)
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means")

plt.show()


#Explanation
#We generated uniform random numbers and repeatedly took samples of size 30. 
#The distribution of sample means becomes approximately normal, demonstrating the Central Limit Theorem.
```
